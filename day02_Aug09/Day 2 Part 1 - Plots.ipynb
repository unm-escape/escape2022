Now let's make a series of values and compute a sin() function to plot.

The **np.arange()** function is like range(), but it creates a numpy array. We give the **np.arrange(start, stop, step values)**.

```python
timeAxis = np.arange(0,10,0.1) # np.arange() is similar to the python range() function that we learned yesterday
# Note that python doesn't print 'zeros'

print(timeAxis)
```

```python
ampSin = np.sin(timeAxis)
print(ampSin)
```

How about a cosine too?

```python
ampCos = np.cos(timeAxis)
print(ampCos)
```

```python
# Plotting them 
plt.figure(1)
# with sine and cosine values in the x-axis
plt.plot(ampSin,timeAxis) # give different line styles

plt.show
```
